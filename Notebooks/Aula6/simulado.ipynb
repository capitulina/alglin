```python
import numpy as np
import autograd.numpy as np_
from autograd import grad
import matplotlib.pyplot as plt

def loss ( parametros_como_tupla ):
    x, y, a, b = parametros_como_tupla
    return np_.mean((y - (a * x + b))**2)

derivada_por_autograd = grad(loss)

a_real = np.random.rand() * 3 + 5
b_real = np.random.rand() * 3 + 5
sigma_real = np.random.rand() * 2
x = np.random.rand(1000)
erro_real = np.random.randn(1000) * sigma_real
y = a_real * x + b_real + erro_real

a_est = 5.0
b_est = 2.0
alpha = 0.01
erros = []
for _ in range(1000):
    erros.append(loss((x, y, a_est, b_est)))
    _, _, da, db = derivada_por_autograd((x, y, a_est, b_est))
    a_est, b_est = a_est - alpha * da, b_est - alpha * db

plt.plot(erros)
print(a_est, b_est)
print(a_real, b_real)
```
