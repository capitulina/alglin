# Demonstração: transformações compostas

```python
import numpy as np
import matplotlib.pyplot as plt

def matriz_de_translacao(dx, dy):
    return np.array([[1, 0, dx],
                     [0, 1, dy],
                     [0, 0, 1]])

def matriz_de_rotacao(theta):
    return np.array([[np.cos(theta), -np.sin(theta), 0],
                      [np.sin(theta), np.cos(theta), 0],
                     [0, 0, 1]])

def nuvem_de_pontos(n_pontos, minx, maxx, miny, maxy):
    pontosx = np.random.uniform(low=minx, high=maxx, size=n_pontos)
    pontosy = np.random.uniform(low=miny, high=maxy, size=n_pontos)
    coordenada_homogenea = np.ones((1, n_pontos))
    
    pontos = np.vstack((pontosx, pontosy))
    return np.vstack((pontos, coordenada_homogenea))
```

```python
X = nuvem_de_pontos(1000, -10, 10, -10, 10)
plt.figure(figsize=(3,3))
plt.scatter(X[0,:], X[1,:], c='b')
plt.show()
```

```python
X = nuvem_de_pontos(1000, -10, 10, -10, 10)
R = matriz_de_rotacao(np.pi/4)
Y = R @ X

plt.figure(figsize=(5,5))
plt.scatter(X[0,:], X[1,:], c='b')
plt.scatter(Y[0,:], Y[1,:], c='r')
plt.show()
```

```python
X = nuvem_de_pontos(1000, -10, 10, -10, 10)
A = matriz_de_translacao(1, 0)
B = matriz_de_translacao(0, 1)
R = matriz_de_rotacao(np.pi/12)

A_ = np.linalg.inv(A)
B_ = np.linalg.inv(B)
```

```python
X = nuvem_de_pontos(1000, -6, -4, -8, -6)
TRANSLACAO = B_ @ B_ @ B_ @ B_ @ B_ @ B_ @ B_  @ A_ @ A_ @ A_ @ A_ @ A_
ROTACAO = R @ R @ R @ R

T = np.linalg.inv(TRANSLACAO) @ ROTACAO @ TRANSLACAO 

Y = T @ X

plt.figure(figsize=(5,5))
plt.scatter(X[0,:], X[1,:], c='b')
plt.scatter(Y[0,:], Y[1,:], c='r')
plt.xlim(-20, 20)  
plt.ylim(-20, 20)
plt.show()
```
