1) 
    - D = 1, determinante = 1
    - D = 1.5, determinante = 1.5
    - D = 2, determinante = 2
    - D = 0, determinante = 0


```python
x = np.random.uniform(-0.5, 0.5, size=(2, 600))
Ds = [1, 1.5, 2, 0]

for D in Ds:
    A = np.array([[1, 0], [0, D]])
    y = A @ x
    det = np.linalg.det(A)
    area_transformada = abs(det)

    plt.figure()
    plt.scatter(x[0,:], x[1,:], s=1, c='r', label='Original')
    plt.scatter(y[0,:], y[1,:], s=1, c='b', label=f'Transformada D={D}')
    plt.xlim([-2, 2])
    plt.ylim([-2, 2])
    plt.grid()
    plt.title(f'D = {D} | Determinante = {det:.2f} | Área transformada = {area_transformada:.2f}')
    plt.legend()
    plt.show()
    print(f"D = {D}: Determinante = {det:.2f}, Área transformada = {area_transformada:.2f}")
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.image as mpimg
from matplotlib.patches import Rectangle
```

Exercicio 1

1) Os eixos representam x, y e numero de canais de cor da imagem.
2) As dimensoes são (427, 640, 3)

Exercicio 2

1) 
    - carrega imagem como array numpy
    - cria uma imagem preta do mesmo tamanho
    - devolve todos os pares do retangula de acordo com a coordenada
    - copia posicao a posicao das coordenadas da primeira imagem na imagem preta

2) 

```python
def criar_indices(min_i, max_i, min_j, max_j):
    import itertools
    L = list(itertools.product(range(min_i, max_i), range(min_j, max_j)))
    idx_i = np.array([e[0] for e in L])
    idx_j = np.array([e[1] for e in L])
    idx = np.vstack( (idx_i, idx_j) )
    return idx
```

```python
image = mpimg.imread("rio_de_janeiro.jpg")
image_ = np.zeros_like(image)

X = criar_indices(0, 300, 0, 640)
image_[X[0,:], X[1,:], :] = image[X[0,:], X[1,:], :]

print(image_.shape)
plt.figure()
plt.imshow(image_)
plt.show()
```

Exercicio 3

```python
image = mpimg.imread("rio_de_janeiro.jpg")
image_ = np.zeros_like(image)

X = criar_indices(0, 300, 0, 300)
Ases = [np.array([[1, 0], [0, 1]]), 
        np.array([[0, 1], [1, 0]]), 
        np.array([[1, 0.2], [0, 1]]), 
        np.array([[1, 0], [0.2, 1]]),
        np.array([[1, 0], [0, 3]])]
for A in Ases:
    Xd = A @ X
    Xd = Xd.astype(int)

    filtro = (
        (Xd[0,:] >= 0) & (Xd[0,:] < image.shape[0]) &
        (Xd[1,:] >= 0) & (Xd[1,:] < image.shape[1])
    )
    Xd_f = Xd[:, filtro]
    X_f = X[:, filtro]

    image_[Xd_f[0,:], Xd_f[1,:], :] = image[X_f[0,:], X_f[1,:], :]


    print(image_.shape)
    plt.figure(figsize=(15,15))
    plt.subplot(1,2,1)
    plt.imshow(image)
    # Adiciona o retângulo indicando a área que foi recortada:
    plt.gca().add_patch(Rectangle((0,0),300,300,linewidth=5,edgecolor='r',facecolor='none'))

    plt.title('Origem')
    plt.subplot(1,2,2)
    plt.imshow(image_)
    plt.title('Destino')
    plt.show()
    image_ = np.zeros_like(image)
```
