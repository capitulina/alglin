```python
import numpy as np
```

```python
fs = 44100

T = 2
t = np.linspace(
    0,
    T,
    fs * T,
    endpoint=False,
)
F0 = 440
x = np.zeros_like(t)
mags = [ 1, 0.5, 1, 0.5, 1, 0.5 ]
for idx, a in enumerate(mags):
    x += a * np.cos( 2 * np.pi * (idx + 1) * F0 * t)

F0 = 440 * 3 / 2
y = np.zeros_like(t)
mags = [ 1, 0.5, 0.25, 0.125, 0.5, .7,  0.8 ]
for idx, a in enumerate(mags):
    y += a * np.cos( 2 * np.pi * (idx + 1) * F0 * t)

z = x + y
```

```python

import matplotlib.pyplot as plt

plt.plot(t, z)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Cosine Wave')
plt.xlim(0, 0.01)
plt.show()
```

```python
from IPython.display import Audio

Audio(z, rate=fs)
```

```python
F = np.abs(np.fft.fft(x))
plt.plot(F)
plt.xlabel('Frequency Bin')
plt.ylabel('Magnitude')
plt.title('Magnitude Spectrum')
plt.show()
```
